## Задание 1

Возможно, вы замечали, что в адресах некоторых страниц (URL) встречаются символы & и =, как на первой строке примера. Это — GET-запрос протокола HTTP, один из способов передать сайту некоторые данные помимо адреса страницы, которую вы желаете открыть. Сайт может использовать эти данные по-разному, например, изменять страницу в зависимости от них.

Пользователи, конечно, не пишут GET-запрос руками, он формируется той страницей, с которой пользователь переходит, нажав на ссылку или кнопку.

Вы же должны написать программу, которая разбирает («парсит», от parse) GET-запрос.

Вопросительный знак ? разделяет в адресе путь к странице, которую мы запрашиваем, от GET-запроса. Нас будет интересовать только запрос. Он состоит из пар ключ-значение; пары отделены друг от друга амперсандом &, а ключ от значения — знаком равенства. В приведенном примере таких пар две, и ключу «text» соответствует значение «котики», а ключу «source» — значение «images_drawing».

Программа должна вывести значение, соответствующее заданному ключу.

В реальности многие символы должны представляться в URL-адресах особым образом, но мы для простоты будем считать, что в ключах и значениях GET-запроса разрешены все цифробуквенные символы, символы пустого пространства (кроме символа перевода строки) и все знаки препинания, кроме ?, &, = и #.

**Формат ввода**

На первой строке вводится URL, содержащий GET-запрос.

На второй строке вводится ключ, присутствующий в этом запросе.

**Формат вывода**

Нужно вывести одну строку — значение, соответствующее ключу.

**Пример**

|Ввод|Вывод|
|---|---|
|`https://yandex.ru/images/search?text=котики&source=images_drawing`<br>`text`|`котики`|


````python
def dict_parse(url,key):
  r = url.split('/')[-1].split('?')[-1].split('&')
  dict_get = dict([el.split('=') for el in r])
  return dict_get[key]

dict_parse("https://yandex.ru/images/search?text=котики&source=images_drawing",'text')
````

## Задание 2

Имеется герой со множеством титулов. Нам нужно красиво записать все титулы героя.

**Формат ввода**

На первой строке вводится число $n$ — количество титулов.

Далее следует $n$ титулов. Каждый титул состоит из одного или нескольких слов, причём слова титула могут быть разбросаны на несколько строк, как в примере (но пустых строк нет). Конец титула отмечен звёздочкой на отдельной строке.

**Формат вывода**

Нужно вывести все титулы в одну, в порядке, обратном вводу, разделяя их запятыми, а слова в пределах одного титула — дефисами.

**Пример**

|Ввод|Вывод|
|---|---|
|3<br>Просто<br>Приятный  Парень<br>*<br>Самый<br>Продуктивный<br>  Выращиватель<br>Бобов<br>*<br>Джек<br>Победитель Великанов<br>*|Джек-Победитель-Великанов, Самый-Продуктивный-Выращиватель-Бобов, Просто-Приятный-Парень|

````python
result = []

for _ in range(int(input())):
  pre_result = []
  while True:
    char = input()
    if char!='*': 
      pre_result.append("-".join(char.split(' ')))
    else:
      break
  result.append('-'.join(pre_result))


print(','.join(result[::-1]))
  


````

**Формат ввода**

Вводятся два числа через пробел – периоды обращения говядины и ключа в днях.
Затем вводится дата их общего старта.
Затем вводится количество дат, которые надо найти.

**Формат вывода**


Выведите заданное количество ближайших дат, не считая введенную, в которые оба тела будут одновременно находиться в точке старта. Если дата попадает на вторник, ее не выводить.

Формат вывода дат: DD Month YY

**Пример 1**

|Ввод|Вывод|
|-|-|
|`2 3`<br>`25/03/2022`<br>`5`|`31 March 22`<br>`06 April 22`<br>`18 April 22`<br>`24 April 22`<br>`30 April 22`|

**Пример 2**

|Ввод|Вывод|
|-|-|
|`4 6`<br>`1/04/2022`<br>`3`|`13 April 22`<br>`25 April 22`<br>`07 May 22`|

**Примечания**

0 – понедельник

````python
import datetime as dt
 
a, b = map(int, input().split()) 
s = input().split('/')
data = dt.date(int(s[2]), int(s[1]), int(s[0]))
n = int(input())
i = 1
 
while i < (n+1):
    m = data + dt.timedelta(days=a*b*i)
    if m.weekday() != 1: 
        print(m.strftime("%d %B %Y"))
    else:
        n += 1
    i += 1
````

Ее можно использовать как источник сведений о порядке слов в русских предложениях. Список называется words, и его можно импортировать из своей программы такой конструкцией:

````python
from swift import words
````

````python
from random import choice
````

````python
def get_dict():
    d = {}
    for i in range(1,len(words)):

        key = words[i-1].lower()
        value = words[i].lower()

        if not d.get(key, False): 
          d[key] = [value] 
        else:  
          d[key].append(''.join(value))
    return d

def generator(dict_words,count):
    counter = 1
    words = [key for key in dict_words.keys()]
    result = []
    while True: 
      x = choice(words)
      y = choice(dict_words[x])

      if y in [',','.',':',';','-']:
        result.append(f"{x}{y}")
      else:
        result.append(f"{x} {y}")

      x = y

      if counter == count:
         break
      counter+=1

    print(''.join(list(map(lambda x: x.capitalize(),' '.join(result).split('.')))))

def main():
    dict_result = get_dict()
    end = int(input('Введите количество слов:\t'))
    generator(dict_result,end)

if __name__ == '__main__':
    main()   
````
